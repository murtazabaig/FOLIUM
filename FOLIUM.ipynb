```python
import folium
my_map = folium.Map(location=map_center, zoom_start=5)

locations = [
    {"location": [15.2993, 74.1240], "popup": " Goa"},
    {"location": [22.5726, 88.3639], "popup": "Kolkata"},
    {"location": [13.0827, 80.2707], "popup": " Chennai"}
]

for loc in locations:
    folium.Marker(
        location=loc["location"],
        popup=loc["popup"],
        icon=folium.Icon(color='red', icon='tree-conifer')  
    ).add_to(my_map)
my_map.save('customized_map_with_locations.html')
my_map
```

```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
file_path = r"C:\Users\HOME\Downloads\state wise centroids_2011.csv"
df = pd.read_csv(file_path)
map_center = [20.5937, 78.9629]
my_map = folium.Map(location=map_center, zoom_start=5)
marker_cluster = MarkerCluster(
    name='Cluster',
    overlay=True,
    control=True,
    icon_create_function=None 
)
for index, row in df.iterrows():
    marker = folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['State']} - {row.get('Population', 'Population not available')}",
        icon=folium.Icon(color='blue', icon='info-sign') 
    )
    marker.add_to(marker_cluster)
marker_cluster.add_to(my_map)
my_map.save('clustered_map_with_state_centroids.html')

my_map
```

```python
#question 5
import folium
import pandas as pd
file_path = r"C:\Users\HOME\Downloads\state wise centroids_2011.csv"
df = pd.read_csv(file_path)
map_center = [20.5937, 78.9629]  
my_map = folium.Map(location=map_center, zoom_start=5)
for index, row in df.iterrows():
    
    population = row.get('YourPopulationColumnName', 0)  
    color = 'green' if population > 5000000 else 'orange'
    
    area_column_name = 'YourAreaColumnName'  
    size = row.get(area_column_name, 0) / 10000  
   
    marker = folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['State']} - Population: {population}, {area_column_name}: {size * 10000}",  
        radius=size,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6
    )

    marker.add_to(my_map)

my_map.save('dynamic_markers_map.html')

my_map
```
